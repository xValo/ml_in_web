```python
from sklearn.datasets import fetch_olivetti_faces
from sklearn.datasets import fetch_rcv1
from sklearn.decomposition import PCA
from sklearn.manifold import SpectralEmbedding, Isomap, LocallyLinearEmbedding
import matplotlib.pyplot as plt
import pandas as pd
import coranking
from coranking.metrics import trustworthiness, continuity, LCMC
```

```python
X, y = fetch_olivetti_faces(return_X_y=True)
isomap = Isomap(n_components=2)
X_iso = isomap.fit_transform(X)
plt.plot(X_iso[:,0], X_iso[:,1], '.')
plt.title('Isomap')
plt.show()

lle = LocallyLinearEmbedding(n_components=2)
X_lle = lle.fit_transform(X)
plt.plot(X_lle[:,0], X_lle[:,1], '.')
plt.title('LLE')
plt.show()

lap_eigenmap = SpectralEmbedding(n_components=2)
X_lap = lap_eigenmap.fit_transform(X)
plt.plot(X_lap[:,0], X_lap[:,1], '.')
plt.title('Laplacian Eigenmap')
plt.show()
```

```python
MAX_K = 50
Q = coranking.coranking_matrix(X, X_iso)
Q1 = coranking.coranking_matrix(X, X_lle)
Q2 = coranking.coranking_matrix(X, X_lap)
# Isomap
trust_iso = trustworthiness(Q, min_k=1, max_k=MAX_K)
cont_iso = continuity(Q, min_k=1, max_k=MAX_K)
lcmc_iso = LCMC(Q, min_k=1, max_k=MAX_K)
# LLE
trust_lle = trustworthiness(Q1, min_k=1, max_k=MAX_K)
cont_lle = continuity(Q1, min_k=1, max_k=MAX_K)
lcmc_lle = LCMC(Q1, min_k=1, max_k=MAX_K)
# Lap
trust_lap = trustworthiness(Q2, min_k=1, max_k=MAX_K)
cont_lap = continuity(Q2, min_k=1, max_k=MAX_K)
lcmc_lap = LCMC(Q2, min_k=1, max_k=MAX_K)

plt.plot(range(1,MAX_K), trust_iso, label='Isomap')
plt.plot(range(1,MAX_K), trust_lle, label='LLE')
plt.plot(range(1,MAX_K), trust_lap, label='Lap')
plt.title('trustworthiness')
plt.xlabel('Neighbours')
plt.legend()
plt.show()

plt.plot(range(1,MAX_K), cont_iso, label='Isomap')
plt.plot(range(1,MAX_K), cont_lle, label='LLE')
plt.plot(range(1,MAX_K), trust_lap, label='Lap')
plt.title('continuity')
plt.xlabel('Neighbours')
plt.legend()
plt.show()

plt.plot(range(1,MAX_K), lcmc_iso, label='Isomap')
plt.plot(range(1,MAX_K), lcmc_lle, label='LLE')
plt.plot(range(1,MAX_K), lcmc_lap, label='Lap')
plt.title('LCMC')
plt.xlabel('Neighbours')
plt.legend()
plt.show()
```
